Q1. How can you create a Bokeh plot using Python code?

You create a Bokeh plot by:

Creating a figure

Adding glyphs (shapes)

Displaying the plot

Example:

```python
from bokeh.plotting import figure, show

p = figure(title="Simple Bokeh Plot", x_axis_label="X", y_axis_label="Y")

x = [1, 2, 3, 4]
y = [2, 5, 3, 6]

p.line(x, y)

show(p)
```

Q2. What are glyphs in Bokeh?

How can you add them? Explain with example.

Glyphs are the basic visual shapes used to represent data in Bokeh plots.

Examples of glyphs:

line

circle

square

triangle

bar

You add glyphs using methods of the figure object.

Example:

```python
from bokeh.plotting import figure, show

x = [1, 2, 3, 4]
y = [3, 6, 2, 5]

p = figure(title="Glyph Example")

p.circle(x, y, size=10, color="blue")
p.line(x, y)

show(p)
```

Q3. How can you customize the appearance of a Bokeh plot

(axes, title and legend)?

You can customize them using properties of the figure object.

Example:

```python
from bokeh.plotting import figure, show

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

p = figure(title="Customized Plot")

p.line(x, y, legend_label="Line Data")

# Title customization
p.title.text_font_size = "16pt"

# Axis labels
p.xaxis.axis_label = "X values"
p.yaxis.axis_label = "Y values"

# Axis appearance
p.xaxis.major_label_text_font_size = "12pt"
p.yaxis.major_label_text_font_size = "12pt"

# Legend position
p.legend.location = "top_left"

show(p)
```

Q4. What is a Bokeh server?

How is it used for real-time interactive plots?

A Bokeh server allows you to run Bokeh applications that:

keep a live connection with the browser

update plots automatically

respond to user interactions

support real-time data updates

In short:

👉 Bokeh server runs Python code on the server and pushes updates to the browser.

Simple example (real-time update):

```python
from bokeh.plotting import figure
from bokeh.server.server import Server
from bokeh.models import ColumnDataSource
from bokeh.layouts import column
import random

def modify_doc(doc):
    source = ColumnDataSource(data=dict(x=[1,2,3], y=[1,2,3]))

    p = figure()
    p.line("x", "y", source=source)

    def update():
        new_y = [random.randint(1,10) for _ in range(3)]
        source.data = dict(x=[1,2,3], y=new_y)

    doc.add_periodic_callback(update, 1000)
    doc.add_root(column(p))

server = Server({'/': modify_doc})
server.start()
```

Q5. How can you embed a Bokeh plot in a web page using Flask or Django?

You can embed a Bokeh plot into a web page by using:

components() from bokeh.embed

Below is a simple Flask example.

Flask example

```python
from flask import Flask, render_template_string
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route("/")
def index():
    p = figure(title="Bokeh in Flask")
    p.line([1,2,3,4], [4,7,2,6])

    script, div = components(p)

    html = """
    <html>
        <head>
            {{ script|safe }}
        </head>
        <body>
            {{ div|safe }}
        </body>
    </html>
    """

    return render_template_string(html, script=script, div=div)

if __name__ == "__main__":
    app.run(debug=True)
```
